# Sentiment Analysis Plotly Charts

This notebook reads the sentiment results CSV and shows interactive Plotly charts.

```python
from pathlib import Path

import pandas as pd
import plotly.express as px
import plotly.io as pio

pio.templates.default = "plotly_dark"
```

```python
PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "Script":
    PROJECT_ROOT = PROJECT_ROOT.parent

RESULTS_PATH = PROJECT_ROOT / "Data" / "event_sentiment_analysis.csv"
sentiment_data = pd.read_csv(RESULTS_PATH)
sentiment_data.head()
```

## Sentiment Label Distribution

```python
sentiment_counts = (
    sentiment_data["Sentiment Label"]
    .value_counts()
    .rename_axis("Sentiment Label")
    .reset_index(name="Number of Events")
)

fig = px.bar(
    sentiment_counts,
    x="Sentiment Label",
    y="Number of Events",
    color="Sentiment Label",
    text="Number of Events",
    color_discrete_map={"Positive": "#2E7D32", "Neutral": "#757575", "Negative": "#C62828"},
    title="Sentiment Label Distribution",
)
fig.update_traces(textposition="outside")
fig.update_layout(
    template="plotly_dark",
    showlegend=False,
    yaxis_title="Events",
    paper_bgcolor="#111827",
    plot_bgcolor="#111827",
)
fig.show()
```

## Sentiment Score by Event

```python
score_by_event = sentiment_data.sort_values("Sentiment Score")

fig = px.bar(
    score_by_event,
    x="Sentiment Score",
    y="Event",
    color="Sentiment Label",
    orientation="h",
    hover_data=["Type", "Registers", "Room Available/Not Available", "Accept/Reject", "Sentiment Reason"],
    color_discrete_map={"Positive": "#2E7D32", "Neutral": "#757575", "Negative": "#C62828"},
    title="Sentiment Score by Event",
)
fig.add_vline(x=0, line_width=1, line_dash="dash", line_color="#424242")
fig.update_layout(
    template="plotly_dark",
    xaxis_range=[-100, 100],
    yaxis_title="",
    height=650,
    paper_bgcolor="#111827",
    plot_bgcolor="#111827",
)
fig.show()
```

## Score Distribution

```python
fig = px.histogram(
    sentiment_data,
    x="Sentiment Score",
    color="Sentiment Label",
    nbins=10,
    color_discrete_map={"Positive": "#2E7D32", "Neutral": "#757575", "Negative": "#C62828"},
    title="Distribution of Sentiment Scores",
)
fig.add_vline(x=0, line_width=1, line_dash="dash", line_color="#424242")
fig.update_layout(
    template="plotly_dark",
    xaxis_range=[-100, 100],
    yaxis_title="Events",
    paper_bgcolor="#111827",
    plot_bgcolor="#111827",
)
fig.show()
```

## Score vs. Registrations

```python
fig = px.scatter(
    sentiment_data,
    x="Registers",
    y="Sentiment Score",
    color="Sentiment Label",
    hover_name="Event",
    hover_data=["Type", "Room Available/Not Available", "Accept/Reject", "Sentiment Reason"],
    color_discrete_map={"Positive": "#2E7D32", "Neutral": "#757575", "Negative": "#C62828"},
    title="Sentiment Score vs. Registration Count",
)
fig.add_hline(y=0, line_width=1, line_dash="dash", line_color="#424242")
fig.update_layout(
    template="plotly_dark",
    yaxis_range=[-100, 100],
    paper_bgcolor="#111827",
    plot_bgcolor="#111827",
)
fig.show()
```
